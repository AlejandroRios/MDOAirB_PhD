```python
import pandas as pd
import os
import plotly.graph_objects as go
```

```python
path = os.getcwd()
print(path)
input = pd.read_csv(path + "/Results/analysis_input_bprlow.csv")
output = pd.read_csv(path + "/Results/analysis_output_bprlow.csv")
```

```python
output
```

```python
fig = go.Figure(go.Carpet(
    a=input["WingArea"],
    b=input["FanDiameter"]/10,
    y=output["fuel_mass"],

    aaxis=dict(
        tickprefix='WA = ',
        ticksuffix='m2',
        smoothing=0.2,
        minorgridcount=10,
    ),
    baxis=dict(
        tickprefix='FD = ',
        ticksuffix='m',
        smoothing=0.4,
        minorgridcount=9,
    )
))

fig.show()
```
