```python
import pandas as pd
import os
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = os.getcwd()
print(path)
input = pd.read_csv(path + "/Results/analysis_single_input_1-new.csv")
output = pd.read_csv(path + "/Results/analysis_single_output_1-new.csv")
```

```python
output.head()
```

```python
distance = output['distance'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
mass = output['mass'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
altitude = output['altitude'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
time = output['time'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
LoD = output['LoD'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
sfc = output['sfc'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
CL = output['CL'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                                .replace('  ', ' '), sep=' '))[0]
CD = output['CD'].apply(lambda x:
                           np.fromstring(
                               x.replace('\n', '')
                                .replace('[', '')
                                .replace(']', '')
                               .replace('  ', ' '), sep=' '))[0]
mach = output['mach.1'].apply(lambda x:
                        np.fromstring(
                            x.replace('\n', '')
                            .replace('[', '')
                            .replace(']', '')
                            .replace('  ', ' '), sep=' '))[0]
thrust = output['thrust'].apply(lambda x:
                              np.fromstring(
                                  x.replace('\n', '')
                                  .replace('[', '')
                                  .replace(']', '')
                                  .replace('  ', ' '), sep=' '))[0]
throttle = output['throttle'].apply(lambda x:
                              np.fromstring(
                                  x.replace('\n', '')
                                  .replace('[', '')
                                  .replace(']', '')
                                  .replace('  ', ' '), sep=' '))[0]
```
